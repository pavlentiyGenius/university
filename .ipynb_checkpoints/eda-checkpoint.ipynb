```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import chain
```

```python
pet_smooth = pd.read_csv('data/PET(smooth).csv')
```

## Show

```python
x = pet_smooth['Raman shift [1/cm]']

plt.figure(figsize=(20, 7))
plt.autoscale()
plt.plot(x, pet_smooth['Intensity'], label='signal', c='b')
```

```python
intensity = np.array(pet_smooth['Intensity'].tolist())
```

```python
intensity
```

```python
pet_smooth = pet_smooth.reset_index(drop=True)
```

## IQR

```python
pet_smooth['outliers'] = False
data = intensity

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

threshold = 2.5 * iqr
outliers = np.where((data < q1 - threshold) | (data > q3 + threshold))

# print("Outliers of array ",data,"is : \n", data[outliers])
```

```python
outliers = outliers[0]
```

```python
pet_smooth.loc[outliers, 'outliers'] = True
```

```python
pet_smooth['outliers'].value_counts()
```

```python
# pet_smooth['color'] = pet_smooth['outliers'].map({True: 'red', False: 'blue'})
```

```python
norm = pet_smooth.copy()
out = pet_smooth.copy()
```

### Interpolate for beauty print

```python
norm['outliers_for_plot'] = norm['outliers']

norm.loc[norm['outliers_for_plot'] == True, 'outliers_for_plot'] = None

# print('Orig outliers: ', norm['outliers_for_plot'].sum())

new_true_indexes = list(set(list(chain.from_iterable([[i-1, i, i+1] for i in list(norm[norm['outliers_for_plot'] == False].index)]))))
new_true_indexes = [i for i in new_true_indexes if i < norm.shape[0]]
new_true_indexes = [i for i in new_true_indexes if i >= 0]

norm.loc[new_true_indexes, 'outliers_for_plot'] = False

print('Interpolated outliers: ', norm['outliers_for_plot'].isna().sum())

norm['outliers_for_plot'] = norm['outliers_for_plot'].fillna(True)
```

```python
norm.loc[norm['outliers_for_plot'] == True, 'Intensity'] = None

print('Norm data: ', norm['Intensity'].shape[0] - norm['Intensity'].isna().sum())
print('Outliers data for print: ', norm['outliers_for_plot'].sum())
```

```python
out.loc[out['outliers'] == False, 'Intensity'] = None

print('Outliers data: ', out['outliers'].sum())
print('Outliers for print data: ', out['outliers'].sum())
```

```python
x = pet_smooth['Raman shift [1/cm]']

plt.figure(figsize=(20, 7))
plt.autoscale()
plt.plot(x, norm['Intensity'],label='signal', c='b', )
plt.plot(x, out['Intensity'],'go--', label='outlier', c='r')
```

```python
import statistics
```

```python
def split_seq(l):
    all_seq = []
    seq = []
    for i in range(len(l) - 1):
        if l[i] + 1 == l[i + 1]:
            seq.append(l[i])
        else:
            all_seq.append(seq)
            seq = []
    seq.append(l[i + 1])
    all_seq.append(seq)
    return all_seq
```

```python
split_out = split_seq(outliers)
```

```python
maxvals = []
for splits in split_out:
    # print(max(pet_smooth.loc[splits]['Raman shift [1/cm]'].tolist()))
    array_shift = np.array(pet_smooth.loc[splits]['Raman shift [1/cm]'].tolist())
    array_intens = np.array(pet_smooth.loc[splits]['Intensity'].tolist())
    argmax = array_intens.argmax()
    x = array_shift[argmax]
    y = array_intens[argmax]
    maxvals.append((x, y))
```

```python
x = pet_smooth['Raman shift [1/cm]']

plt.figure(figsize=(20, 7))
plt.autoscale()
plt.plot(x, norm['Intensity'], label='signal', c='b')
plt.plot(x, out['Intensity'], label='outlier', c='r')
for i in maxvals:
    plt.scatter(i[0], i[1], c='black', s=20)
```

```python
ouliers_df = pd.DataFrame(maxvals, columns=['Raman shift [1/cm]_outlier', 'Intensity_outlier'])
```
